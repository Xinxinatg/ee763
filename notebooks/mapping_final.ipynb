```python
import os
import pandas as pd
import glob
originalfolder = "/home/debanjan/Desktop/ee763/ee_dataset/annotations/"
destinationfolder = "/home/debanjan/Desktop/ee763/csv_annotations/"
```

```python
if not os.path.exists(destinationfolder):
    print("create the destination folder if not exist")
    os.makedirs(destinationfolder)
```

```python
def get_name(src):
    lst = src.split('/')
    res = []
    for x in lst: 
        if x: res.append(x)
    n = len(res)
    return res[n-3]+'_'+res[n-2]
```

```python
print(get_name('/home/debanjan/Desktop/ee763/ee_dataset/annotations/bookstore/video0/annotations.txt'))
```

```python
def csv_mapping(src,dest):
    print("-"*50)
    print(src)
    print(dest)
    f = open(src,'r')
    data = f.read()
    data = data.replace(' ',',')
    dest = dest+'annotations.csv'
    f = open(dest,'w')
    f.write(data)
    f.close()
    ann = pd.read_csv(dest,names=['track_id','xmin','ymin','xmax','ymax','frame','lost','occluded','generated','label'])
    ann = ann.drop('track_id', 1)
    ann = ann.drop('lost', 1)
    ann = ann.drop('occluded', 1)
    ann = ann.drop('generated', 1)
    df = ann.loc[ann['frame']==0]
    length = len(ann)
    for i in range(40,length,150):
        x = ann.loc[ann['frame']==i]
        df = x.append(df,ignore_index=True)
    length = len(df['frame'])
    for i in range(length):
        filename = get_name(src)
        print('imgs/train/'+filename+'_'+str(df['frame'][i])+'.png')
        df['frame'][i] = 'imgs/train/'+filename+'_'+str(df['frame'][i])+'.png'
    lst = src.split('/')
    res = []
    for x in lst: 
        if x: res.append(x)
    n = len(res)
    df.to_csv('train'+res[n-3]+'.csv',sep='\t',index=False)
    print("Done! Converted to CSV.")
    print("-"*50)
```

```python
subfolders_path = sorted(glob.glob(originalfolder+"*"))
```

```python
subfolders = []
for i in range(len(subfolders_path)):
    subfolder = os.path.basename(subfolders_path[i])
    subfolders.append(subfolder)
    subsubfolders = []
    subsubfolders_path = glob.glob(originalfolder+subfolders[-1]+"/"+"*")
    for j in range(len(subsubfolders_path)):
        subsubfolders.append(os.path.basename(sorted(subsubfolders_path)[j]))
    if not os.path.exists(destinationfolder+subfolders[i]+"/"):
        print("creating the subfolder..."+destinationfolder+subfolders[i]+"/")
        os.makedirs(destinationfolder+subfolders[i]+"/")
        for j in range(len(subsubfolders)):
            if not os.path.exists(destinationfolder+subfolders[i]+"/"+subsubfolders[j]+"/"):
                print("Creating subsubfoleder ..."+destinationfolder+subfolders[i]+"/"+subsubfolders[j]+"/")
                os.makedirs(destinationfolder+subfolders[i]+"/"+subsubfolders[j]+"/")
    for j in range(len(subsubfolders)):
        print ("Now proceeding " + subsubfolders[j])
        csv_mapping(originalfolder+subfolders[i]+"/"+subsubfolders[j]+"/"+"annotations.txt", 
            destinationfolder+subfolders[i]+"/"+subsubfolders[j]+"/")
```
